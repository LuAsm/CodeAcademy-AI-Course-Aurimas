5. Write a Pandas program to merge two given dataframes with different columns. <br>
Data: data1 & data2


|| key1 | key2 | P | Q |
| --- | --- | --- | --- | --- |
0 | K0 | K0 | P0 | Q0 |
0 | K0 | K1 | P1 | Q1 |
0 | K1 | K0 | P2 | Q2 |
0 | K2 | K1 | P3 | Q3 |


|| key1 | key2 | R | S |
| --- | --- | --- | --- | --- |
0 | K0 | K0 | R0 | S0 |
0 | K1 | K0 | R1 | S1 |
0 | K1 | K0 | R2 | S2 |
0 | K2 | K0 | R3 | S3 |

```python
import pandas as pd
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})
result = pd.concat([data1, data2], axis=0, ignore_index=True)
result
```
