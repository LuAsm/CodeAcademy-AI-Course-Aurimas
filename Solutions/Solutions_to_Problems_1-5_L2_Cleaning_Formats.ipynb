3. You developed a multiplayer indie game called 🌹 Rose Thorn. Players compete in one of two venues - the ocean or the desert. You track the outcome of five games between three players in a DataFrame called games.

```python
import numpy as np
import pandas as pd

games = pd.DataFrame({
    'bella1':   ['2nd', '3rd', '1st', '2nd', '3rd'],
    'billybob': ['1st', '2nd', '2nd', '1st', '2nd'],
    'nosoup4u': ['3rd', '1st', '3rd', '3rd', '3rd'],
    'venue': ['desert', 'ocean', 'desert', 'ocean', 'desert']
})

games
```

Now you want to analyze the data. Convert the `games` DataFrame into a new DataFrame that identifies how many times each (`player`, `placement`) occurs per `venue`, specifically with `venue` as the row index and (`player`, `placed`) as the column MultiIndex.

```python
# 1. The first step is to use the DataFrame.melt() function to "stack" the columns bella1, billybob, and nosoup4u into a single column.

step1 = games.melt(id_vars='venue', var_name='player', value_name='placed')
step1
```

```python
# 2. Next, if we group the data by (venue, player, placed), then aggregate the groups with 'size', we get back the count of rows in each group. This tells us how many times each player got each placement for each venue, amongst the triplets that actually occurred in the data.

step2 = step1.groupby(['venue', 'player', 'placed']).agg('size')
step2
```

```python
# 3. However, we also want triplets that didn't occur, like 
#(desert, billybob, 3rd) and (ocean, nosoup4u, 2nd). 
#There's a few ways can fix this, but one really cool way is, 
#let's got back to step1 convert 'venue', 'player', and 'placed' to Categoricals.

step1['venue'] = pd.Categorical(step1.venue)
step1['player'] = pd.Categorical(step1.player)
step1['placed'] = pd.Categorical(step1.placed)
```

```python
# 4. Now when we group the data in step2, all those unobserved triplets are included in the result!

step2 = step1.groupby(['venue', 'player', 'placed']).agg('size')
step2
```

```python
# 5. At this point, step2 is a dataframe with a 3-level multiindex. 
#If we reshape it using unstack(), with level=['player','placed'] 
#to move the player and placed levels out of the row index and 
#into the column index, we get back our desired result.

step3 = step2.unstack(level=['player', 'placed'])
step3
```

4. You run an ecommerce site called shoesfordogs.com . You want to analyze your visitors, so you compile a DataFrame called hits that represents each time a visitor hit some page on your site.

```python
import numpy as np
import pandas as pd

# exposed places
exposed = [0,5,9]

# whereabouts of each person
generator = np.random.default_rng(2468)
Nplaces = 10
Npersons = 10
place_ids = np.arange(Nplaces)
visits = generator.choice(place_ids, size=3*Nplaces, replace=True)
split_idxs = np.sort(generator.choice(len(visits), size=9, replace=True))
whereabouts = pd.DataFrame({
    'person_id': range(Npersons),
    'places': [np.unique(x).tolist() for x in np.array_split(visits, split_idxs)]
})

whereabouts
```

For each `person`, identify the `places` they visited which have been `exposed`. Make this a new list-column in `whereabouts` called `exposures`.


```python
#SOLUTION 1
whereabouts['exposures'] = whereabouts['places'].apply(lambda x: list(set(x) & set(exposed)))
whereabouts
```

```python
# SOLUTION 2
# 1. Expand whereabouts. We use the DataFrame.explode() method with column='places'.
expanded = whereabouts.explode(column='places')
expanded
```

```python
# 2. Select rows where places is one of the exposed places.
filtered = expanded.loc[expanded.places.isin(exposed)]
filtered
```

```python
# 3. Group the data by person_id, aggregating the exposed places with a list.
aggregated = filtered.groupby('person_id')[['places']].agg(list)
aggregated
```

```python
# 4. Rename the column 'places' to 'exposures'.
aggregated.rename(columns={'places': 'exposures'}, inplace=True)
aggregated
```

```python
# 5. Insert this column of exposures into the whereabouts DataFrame. A left join does the trick.
whereabouts = pd.merge(left=whereabouts, right=aggregated, how='left', on='person_id')
whereabouts
```

```python
# 6. Replace NaNs with empty lists.
# Unfortunately you can't use fillna() to fill NaNs with an empty list. So, we have to do something like this..
# Here, lambda x: returns x if it's already a list, otherwise it returns an empty list [].
whereabouts['exposures'] = whereabouts.exposures.apply(lambda x: x if isinstance(x, list) else [])
whereabouts
```

5. You run an ecommerce site called shoesfordogs.com . You want to analyze your visitors, so you compile a DataFrame called hits that represents each time a visitor hit some page on your site. 

```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

generator = np.random.default_rng(90)
products = ['iev','pys','vae','dah','yck','axl','apx','evu','wqv','tfg','aur','rgy','kef','lzj','kiz','oma']
hits = pd.DataFrame({
    'visitor_id': generator.choice(5, size=20, replace=True) + 1,
    'session_id': generator.choice(4, size=20, replace=True),
    'date_time': pd.to_datetime('2020-01-01') + pd.to_timedelta(generator.choice(60, size=20), unit='m'),
    'page_url': [f'shoesfordogs.com/product/{x}' for x in generator.choice(products, size=20, replace=True)]
})
hits['session_id'] = hits.visitor_id * 100 + hits.session_id
```

You suspect that the undocumented third-party tracking system on your website is buggy and sometimes splits one session into two or more session_ids. You want to correct this behavior by creating a field called session_group_id that stitches broken session_ids together.

Two session, A & B, should belong to the same session group if <br>
 - They have the same visitor_id and <br>
    - Their hits overlap in time or <br>
    - The latest hit from A is within five minutes of the earliest hit from B, or vice-versa <br>

<b>Associativity applies.</b> So, if A is grouped with B, and B is grouped with C, then A should be grouped with C as well.

Create a column in hits called session_group_id that identifies which hits belong to the same session group.

```python
hits.sort_values(by=['visitor_id', 'date_time'], inplace=True)
hits
```

```python
sessions = hits.groupby(['visitor_id', 'session_id']).agg(
    date_time_min=('date_time', 'min'),
    date_time_max=('date_time', 'max')
)
sessions
```

```python
sessions.sort_values(['visitor_id', 'date_time_min'], inplace=True)
sessions
```

```python
sessions['date_time_max_p5'] = sessions.date_time_max + pd.Timedelta(minutes=5)
sessions
```

```python
sessions['date_time_max_p5_cummax'] = sessions.groupby('visitor_id')['date_time_max_p5'].cummax()
sessions
```

```python
sessions['date_time_max_p5_cummax_prev'] = sessions.groupby('visitor_id')['date_time_max_p5_cummax'].transform(pd.Series.shift)
sessions
```

```python
sessions['group_with_prev'] = sessions.date_time_min <= sessions.date_time_max_p5_cummax_prev
sessions
```

```python
sessions['sessions_group_id'] = (~sessions.group_with_prev).cumsum()
sessions
```

```python
hits = pd.merge(left=hits, right=sessions[['sessions_group_id']], on='session_id')
hits
```
