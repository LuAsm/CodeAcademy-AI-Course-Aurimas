1. Thirty college students were surveyed on

- the number of derivatives they've calculated in the past year and
- the number of dates they've been on in the past year


```python
import random

random.seed(1)
derivatives = [random.randrange(0, 250) for i in range(30)]
dates = [random.randrange(0, max(50 - d, 3)) for d in derivatives]

print(derivatives)
# [34, 145, 216, ..., 178, 114, 68]

print(dates)
# [7, 2, 0, ..., 0, 1, 2]
```

2. You own a chain of drive-thru daiquiris 🥤. (Yes, they exist.) Build a plot like the one below showing the distribution of sales per store, per hour.

```python
import numpy as np

rng = np.random.default_rng(1234)

p1 = [0.05, 0.03, 0.02, 0.01, 0.01, 0.01, 0.01, 0.01, 0.02, 0.02, 0.03, 0.04,
      0.07, 0.07, 0.08, 0.08, 0.08, 0.12, 0.15, 0.18, 0.15, 0.12, 0.08, 0.06]
p2 = np.concatenate((p1[-5:], p1[:-5]))
p3 = np.concatenate((p1[5:], p1[:5]))

# sales per store
s1 = rng.choice(24, size=1000, replace=True, p=p1/np.sum(p1))
s2 = rng.choice(24, size=2000, replace=True, p=p2/np.sum(p2))
s3 = rng.choice(24, size=1750, replace=True, p=p3/np.sum(p3))

print(s1[:5]) # [23 16 22 13 14] 
print(s2[:5]) # [22 17 11 22  0]
print(s3[:5]) # [16  7 12 15 17]
```

Plot the image as shown below. (Hint you will need tight_layout() and subplots_adjust() at the end)

```python
import matplotlib.pyplot as plt

sales = [s1, s2, s3]
colors = ['tab:blue', 'tab:orange', 'tab:green']
fig, axs = plt.subplots(3, figsize=(10, 6), sharex=True, gridspec_kw={'hspace':0})
```

```python
for i, ax in enumerate(axs):
    ax.hist(
        x=sales[i],
        bins=24,
        color=colors[i],
        edgecolor='white',
        label=f'store {i+1}'
    )
```

```python
ax.set_xticks(
    ticks=np.arange(0, 24, 2),
    labels=np.arange(0, 24, 2)
)
```

```python
fig.legend(loc='lower center', ncol=3)
```

```python
fig.suptitle('Daiquiris sold per store, per hour', fontweight='bold')
```

```python
fig.tight_layout()
fig.subplots_adjust(bottom=0.11)
```

```python
import matplotlib.pyplot as plt

sales = [s1, s2, s3]
colors = ['tab:blue', 'tab:orange', 'tab:green']
fig, axs = plt.subplots(3, figsize=(10, 6), sharex=True, gridspec_kw={'hspace':0})

for i, ax in enumerate(axs):
    ax.hist(
        x=sales[i],
        bins=24,
        color=colors[i],
        edgecolor='white',
        label=f'store {i+1}'
    )
    
ax.set_xticks(
    ticks=np.arange(0, 24, 2),
    labels=np.arange(0, 24, 2)
)

fig.legend(loc='lower center', ncol=3)

fig.suptitle('Daiquiris sold per store, per hour', fontweight='bold')
fig.tight_layout()
fig.subplots_adjust(bottom=0.11)
```
